```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/tarun/Downloads/Titanic/tested.csv')

print(df.head())

plt.figure(figsize=(10,6))
plt.hist(df['Age'], bins=30, color='lightblue', edgecolor='black')
plt.title('Histogram for Age of people in Titanic')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('C:/Users/tarun/Downloads/Titanic/tested.csv')

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5, color='blue', edgecolor='black')

plt.title('Scatter plot of Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(alpha=0.75)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/tarun/Downloads/Titanic/tested.csv')

plt.figure(figsize=(10, 6))
df.boxplot(column='Fare', by='Pclass', grid = False)

plt.title('Box Plot of fare by passenger class')
plt.suptitle('')
plt.xlabel('passenger class')
plt.ylabel('Fare')
plt.show()
```
